```python
happiness_pivot_final = happiness.pivot(index="Country name", columns="Year", values=["Generosity", "Freedom to make life choices", "Positive affect", "Democratic Quality", "Life Ladder", "Log GDP per capita", "Social support"])
```

```python
happiness_pivot_final[:5]
```

```python
beer_c = beer_pivot.copy()
```

```python
beer_c = beer_c.T.reset_index()
mindex = ["beer"] * len(beer_c)
beer_c.index = mindex
beer_pivot_final = beer_c.reset_index().set_index(["index", "Year"]).T
```

```python
beer_pivot_final[:5]
```

```python
pop_c = pop.set_index("Country Name").drop(["Country Code", "Indicator Name", "Indicator Code"], axis=1).copy()
```

```python
pop_c = pop_c.T.reset_index()
mindex = ["population"] * len(pop_c)
pop_c.index = mindex
pop_pivot_final = pop_c.reset_index().set_index(["level_0", "index"]).T
```

```python
pop_pivot_final[:5]
```

```python
happiness_pivot_final[:5]
```

```python
beer_pivot_final[:5]
```

```python
bhp = beer_pivot_final.merge(pop_pivot_final, left_index=True, right_index=True).merge(happiness_pivot_final, left_index=True, right_index=True)
bhp[:5]
```
